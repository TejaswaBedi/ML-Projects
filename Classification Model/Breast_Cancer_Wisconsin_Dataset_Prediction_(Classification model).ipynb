# Breast  Cancer (Wisconsin Dataset) Prediction (Classification)

# Importing Libraries

```python
import numpy as np
import pandas as pd
```

# Loading Dataset

```python
dataset=pd.read_csv("Breast_Cancer_Prediction.csv")
dataset = dataset.loc[:, ~dataset.columns.str.contains('^Unnamed')]
print(dataset.shape)
dataset.head(10)
```

```python
dataset.info()
```

# Dropping Unneccesary Features

```python
dataset.drop(['id'],axis=1,inplace=True)
dataset.head(10)
```

```python
dataset.info()
```

# Mapping of Categorical Data

```python
dict={'M':0,'B':1}
dataset1=dataset
dataset1['diagnosis']=dataset1['diagnosis'].map(dict)
print(dataset1['diagnosis'].value_counts())
dataset1.head(20)
```

# Checking for Missing Values in dataset

```python
pd.set_option('display.max_rows',None)
```

```python
print(dataset1.isnull().sum())
```

# Sepearing Training Parameters from Target Value 

```python
x=dataset1.iloc[:,1:31]
print(x)
y=dataset1.iloc[:,0:1]
print(y)
```

# Splitting Data for Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=0)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

# Feature Scaling 

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

# Model 1 - Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
logi=LogisticRegression(penalty='l1',C=0.1,solver='liblinear',multi_class='ovr')
logi.fit(x_train,y_train)
y_train_pred=logi.predict(x_train)
y_test_pred=logi.predict(x_test)
#print(np.concatenate((y_test_pred.reshape(len(y_test_pred),1),y_test.reshape(len(y_test),1)),1))
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
logi_train_acc=accuracy_score(y_train,y_train_pred)
logi_test_acc=accuracy_score(y_test,y_test_pred)
print("Training Accuracy : ",(logi_train_acc)*100,"%")
print("Testing Accuracy :  ",logi_test_acc*100,"%")
print("Confudion Matrix\n")
cm=confusion_matrix(y_test,y_test_pred)
print(cm)
```

# Model 2 - SVM

```python
from sklearn.svm import SVC
svc1=SVC(kernel='rbf',C=10)
svc1.fit(x_train,y_train)
y_train_pred=svc1.predict(x_train)
y_test_pred=svc1.predict(x_test)
#print(np.concatenate((y_test_pred.reshape(len(y_test_pred).1),y_test.reshape(len(y_test),1)),1))
```

```python
svc_train_acc=accuracy_score(y_train,y_train_pred)
svc_test_acc=accuracy_score(y_test,y_test_pred)
print("Training Accuracy : ",svc_train_acc*100,"%")
print("Testing Accuracy :  ",svc_test_acc*100,"%")
print("\nConfusion Matrix\n")
cm=confusion_matrix(y_test,y_test_pred)
print(cm)
```

# Model 3 - KNN

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=15,weights='uniform',p=2)
knn.fit(x_train,y_train)
y_train_pred=knn.predict(x_train)
y_test_pred=knn.predict(x_test)
```

```python
knn_train_acc=accuracy_score(y_train,y_train_pred)
knn_test_acc=accuracy_score(y_test,y_test_pred)
print("Training Accuracy : ",knn_train_acc*100,"%")
print("Testing Accuracy :  ",knn_test_acc*100,"%")
print("\nConfusion Matrix\n")
cm=confusion_matrix(y_test,y_test_pred)
print(cm)
```

# Model 4 - Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB
nb=GaussianNB()
nb.fit(x_train,y_train)
y_train_pred=nb.predict(x_train)
y_test_pred=nb.predict(x_test)

# Naive Bayes has no parameter to tune therefore no hyper-parameter tuning can be performed
```

```python
ACCURACY of Model 4 - Naive Bayes
```

```python
nb_train_acc=accuracy_score(y_train,y_train_pred)
nb_test_acc=accuracy_score(y_test,y_test_pred)
print("Training Accuracy : ",nb_train_acc*100,"%")
print("Testing Accuracy :  ",nb_test_acc*100,"%")
print("\nConfusion Matrix\n")
cm=confusion_matrix(y_test,y_test_pred)
print(cm)
```
